```
def group_and_barplot(feat_groupby, feat_agg='price_value', agg_func=[np.mean, np.size, np.median], feat_sort='size', top=50, subplots=True):
  return (
    df
    .groupby(feat_groupby)[feat_agg]
    .agg(agg_func)
    .sort_values(by=feat_sort, ascending=False)  
    .head(top)
  ).plot(kind='bar', figsize=(20,5), subplots=subplots) 
```
